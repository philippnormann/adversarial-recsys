```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import tikzplotlib
```

```python
plt.style.use('seaborn')
```

```python
x = np.arange(-4.0, 4.0, 0.1)
sigmoid = 1/(1 + np.exp(-x)) 
tanh = np.tanh(x)
relu = np.maximum(x, 0)
plt.plot(x, sigmoid, "-", lw=4, label='σ(x)')
plt.plot(x, tanh, "-", lw=4, label='tanh(x)')
plt.plot(x, relu, "-", lw=4, label='ReLu(x)')
plt.legend(loc="upper left")
plt.ylim(-1, 1)
tikzplotlib.save('../results/figures/activations.tex')
```
